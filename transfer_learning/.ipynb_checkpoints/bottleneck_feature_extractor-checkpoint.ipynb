```python
import torch
import torchvision.models as models
from torchvision import transforms
from PIL import Image

import numpy as np
import matplotlib.pyplot as plt



# please download this sample file
# for example on linux
# wget https://en.wikipedia.org/wiki/File:African_Bush_Elephant.jpg
img_path = 'elephant.jpg'

with open(img_path, 'rb') as f:
  with Image.open(f) as img:
    img = img.convert('RGB')


```

```python
plt.imshow(img)
```

```python
img.size
```

```python
model = models.resnet18(pretrained=True) 
```

```python
model.eval()
```

```python
transform = transforms.Compose([                                                
                               transforms.Resize([224,224]),                    
                               transforms.ToTensor(),                           
                               transforms.Normalize(mean=[0.485, 0.456, 0.406], 
                                        std=[0.229, 0.224, 0.225])              
                               ])   
```

```python
input_tensor = transform(img)         # 3x330x495 -> 3x244x24 size may differ  
input_tensor = input_tensor.unsqueeze(0) # 3x244x2244 -> 1x3x244x244  
```

```python
input = torch.autograd.Variable(input_tensor,                                   
    requires_grad=False)  
```

```python
output_logits = model(input)  
```

```python
_, preds = torch.max(output_logits, 1)  
```

```python
input_tensor.shape
```

```python
_, preds = torch.topk(output_logits, 3)
```
